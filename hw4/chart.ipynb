```python
# coding=utf-8
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator
import numpy as np
```

```python
block_size = [8, 16, 24, 32]
no_share = [51.14, 50.55, 59.62, 49.78]
share = [22.73, 11.13, 9.54, 9.18]

fig, ax = plt.subplots()
x_major_locator=MultipleLocator(8)
ax.xaxis.set_major_locator(x_major_locator)
ax.plot(block_size,
        no_share, 
        'c*-',
        linestyle = '-',
        label = 'w/o shared memory'
        )
ax.plot(block_size,
        share, 
        'm.-.',
        linestyle = '-',
#         color='blue',
        label = 'w/ shared memory'
        )
            
ax.set_xlabel('block_size', fontsize=14)
ax.set_ylabel('time (s)', fontsize=14)
```
